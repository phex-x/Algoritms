1) При помощи стека (можно использовать любую реализацию стека, например, из лекции) проверить, что в строке, содержащей большой фрагмент кода на python, корректно открываются и закрываются скобки '(' и ')'. Заводить какие-либо счетчики и вообще вести подсчет скобок запрещено.

```python
from collections import deque


def is_brackets_balanced(code):
  """
  Проверяет, корректно ли открываются и закрываются скобки '(' и ')' в строке.

  Args:
    code: str: Строка с кодом.

  Returns:
    bool: True, если скобки корректно расставлены, False - в противном случае.
  """
  stack = deque()
  for char in code:
    if char == '(':
      stack.append(char)
    elif char == ')':
      if not stack or stack.pop() != '(':
        return False
  return not stack


if __name__ == "__main__":
  code = """
  def my_function(a, b):
    if a > b:
      return a
    else:
      return b
  """
  print(is_brackets_balanced(code))

  code = """
  def my_function(a, b):
    if a > b:
      return a
    else:
      return b)
  """
  print(is_brackets_balanced(code))
```

2) С помощью списка (list) реализовать стек. При создании стека нужно передать максимальную длину стека. В случае попытки добавления элемента, приводящего к превышению емкости стека возвращать ошибку.
Стек должен поддерживать следующие операции:
S.push(e): добавление элемента e на вершину стека S.
S.pop(): удаляет и возвращает верхний элемент стека S. Если стек пуст, то возникает ошибка.
S.top(): возвращает верхний элемент стека S, не удаляя его. Если стек пуст, то возникает ошибка.
S.is_empty( ): возвращает True если стек S не содержит ни одного элемента.
len(S): возвращает текущеее количество элементов в стеке S.


```python
Реализация стека с помощью списка
Python
class Stack:
  """
  Стек, реализованный с помощью списка.

  Args:
    max_size: int: Максимальная длина стека.
  """

  def __init__(self, max_size):
    self.max_size = max_size
    self.stack = []

  def push(self, element):
    """
    Добавляет элемент на вершину стека.

    Args:
      element: any: Добавляемый элемент.

    Raises:
      StackOverflowError: Если стек переполнен.
    """
    if len(self.stack) >= self.max_size:
      raise StackOverflowError("Стек переполнен.")
    self.stack.append(element)

  def pop(self):
    """
    Удаляет и возвращает верхний элемент стека.

    Raises:
      StackUnderflowError: Если стек пуст.
    """
    if len(self.stack) == 0:
      raise StackUnderflowError("Стек пуст.")
    return self.stack.pop()

  def top(self):
    """
    Возвращает верхний элемент стека, не удаляя его.

    Raises:
      StackUnderflowError: Если стек пуст.
    """
    if len(self.stack) == 0:
      raise StackUnderflowError("Стек пуст.")
    return self.stack[-1]

  def is_empty(self):
    """
    Проверяет, пуст ли стек.

    Returns:
      bool: True, если стек пуст, False - в противном случае.
    """
    return len(self.stack) == 0

  def __len__(self):
    """
    Возвращает текущее количество элементов в стеке.

    Returns:
      int: Количество элементов в стеке.
    """
    return len(self.stack)


if __name__ == "__main__":
  stack = Stack(5)
  stack.push(1)
  stack.push(2)
  stack.push(3)
  print(stack.top())  # 3
  stack.pop()
  stack.push(4)
  stack.push(5)
  print(len(stack))  # 5
  stack.push(6)  # StackOverflowError

  stack = Stack(0)
  stack.pop()  # StackUnderflowError
```

3) Реализовать класс однонаправленного связанного списка.

3.1) Реализовать метод добавления элемента в начало списка.


```python
class Node:
  """
  Узел однонаправленного связанного списка.
  """

  def __init__(self, value, next_node=None):
    self.value = value
    self.next_node = next_node


class LinkedList:
  """
  Однонаправленный связанный список.
  """

  def __init__(self):
    self.head = None

  def add_to_front(self, value):
    """
    Добавляет элемент в начало списка.

    Args:
      value: any: Добавляемое значение.
    """
    new_node = Node(value, self.head)
    self.head = new_node

  def __len__(self):
    """
    Возвращает количество элементов в списке.

    Returns:
      int: Количество элементов.
    """
    count = 0
    current_node = self.head
    while current_node:
      count += 1
      current_node = current_node.next_node
    return count

  def __str__(self):
    """
    Возвращает строковое представление списка.

    Returns:
      str: Строка, представляющая список.
    """
    values = []
    current_node = self.head
    while current_node:
      values.append(str(current_node.value))
      current_node = current_node.next_node
    return "[" + ", ".join(values) + "]"


if __name__ == "__main__":
  linked_list = LinkedList()
  linked_list.add_to_front(10)
  linked_list.add_to_front(20)
  linked_list.add_to_front(30)
  print(linked_list)  # [30, 20, 10]
  print(len(linked_list))  # 3
```

4) На базе класса однонаправленного связанного списка реализовать двунаправленный связанный список.

4.1) Реализовать метод добавления элемента в начало списка.


```python
class Node:
  """
  Узел однонаправленного связанного списка, дополненный ссылкой на предыдущий элемент.
  """

  def __init__(self, value, next_node=None, prev_node=None):
    self.value = value
    self.next_node = next_node
    self.prev_node = prev_node


class DoublyLinkedList:
  """
  Двунаправленный связанный список, реализованный на базе однонаправленного.
  """

  def __init__(self):
    self.head = None
    self.tail = None

  def add_to_front(self, value):
    """
    Добавляет элемент в начало списка.

    Args:
      value: any: Добавляемое значение.
    """
    new_node = Node(value, self.head)
    if self.head:
      self.head.prev_node = new_node
    self.head = new_node
    if not self.tail:
      self.tail = new_node

  def __len__(self):
    """
    Возвращает количество элементов в списке.

    Returns:
      int: Количество элементов.
    """
    count = 0
    current_node = self.head
    while current_node:
      count += 1
      current_node = current_node.next_node
    return count

  def __str__(self):
    """
    Возвращает строковое представление списка.

    Returns:
      str: Строка, представляющая список.
    """
    values = []
    current_node = self.head
    while current_node:
      values.append(str(current_node.value))
      current_node = current_node.next_node
    return "[" + ", ".join(values) + "]"


if __name__ == "__main__":
  linked_list = DoublyLinkedList()
  linked_list.add_to_front(10)
  linked_list.add_to_front(20)
  linked_list.add_to_front(30)
  print(linked_list)  # [30, 20, 10]
  print(len(linked_list))  # 3
```

5) При помощи класса связанного списка реализовать очередь.

```python
class Node:
  """
  Узел однонаправленного связанного списка.
  """

  def __init__(self, value, next_node=None):
    self.value = value
    self.next_node = next_node


class Queue:
  """
  Очередь, реализованная на базе однонаправленного связанного списка.
  """

  def __init__(self):
    self.head = None
    self.tail = None

  def enqueue(self, value):
    """
    Добавляет элемент в конец очереди.

    Args:
      value: any: Добавляемое значение.
    """
    new_node = Node(value)
    if self.tail:
      self.tail.next_node = new_node
    self.tail = new_node
    if not self.head:
      self.head = new_node

  def dequeue(self):
    """
    Удаляет и возвращает элемент из начала очереди.

    Raises:
      QueueUnderflowError: Если очередь пуста.
    """
    if not self.head:
      raise QueueUnderflowError("Очередь пуста.")
    value = self.head.value
    self.head = self.head.next_node
    if not self.head:
      self.tail = None
    return value

  def is_empty(self):
    """
    Проверяет, пуста ли очередь.

    Returns:
      bool: True, если очередь пуста, False - в противном случае.
    """
    return self.head is None

  def __len__(self):
    """
    Возвращает количество элементов в очереди.

    Returns:
      int: Количество элементов.
    """
    count = 0
    current_node = self.head
    while current_node:
      count += 1
      current_node = current_node.next_node
    return count


if __name__ == "__main__":
  queue = Queue()
  queue.enqueue(1)
  queue.enqueue(2)
  queue.enqueue(3)
  print(queue)  # [1, 2, 3]
  print(len(queue))  # 3
  print(queue.dequeue())  # 1
  print(queue.dequeue())  # 2
  print(queue.dequeue())  # 3
  print(queue.is_empty())  # True
  
```
